```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Start coding!
```

```python
# Loading netflix_data.csv file and storing as netflix_df
netflix_df = pd.read_csv('netflix_data.csv')
print(netflix_df.head())
```

```python
# Filtering the data to remove TV shows and store as netflix_subset
netflix_subset = netflix_df[netflix_df['type'] != "TV Show"]
```

```python
# Subsetting and reducing num of columns of netflix movies and saving as netflix_movies
netflix_movies = netflix_subset[['title', 'country', 'genre', 'release_year', 'duration']]
```

```python
# Movies that are shorter than 60 min saving as short_movies
short_movies = netflix_movies[netflix_movies["duration"] < 60 ]
print(short_movies["genre"].value_counts().sort_values(ascending=False))
```

```python
# Initialize colors list
colors = []

for label,row in netflix_movies.iterrows():
    if row["genre"] == "Children":
        colors.append("pink")
    elif row["genre"] == "Documentaries":
        colors.append("blue")
    elif row["genre"] == "Stand-Up":
        colors.append("red")
    else:
        colors.append("grey")

# Initialize figure
fig = plt.figure()
plt.scatter(x=netflix_movies.release_year, y = netflix_movies.duration, c = colors, )
plt.xlabel("Release year")
plt.ylabel("Duration (min)")
plt.title("Movie Duration by Year of Release")
```
